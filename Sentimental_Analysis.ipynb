```python
# importing the usefull libraries
import pandas as pd
import plotly.express as px
from textblob import TextBlob
from wordcloud import WordCloud,STOPWORDS
```

```python
# Reading the Dataset
Sentiment_df=pd.read_csv('C:\\Users\\CH.BHOGESH VMSN\\OneDrive\\Desktop\\Sentimental_Analysis\\DataSet\\mobile_reviews_Mock_Data.csv')
```

```python
#Deriving the polarity of the text
TextBlob('india is a good country').sentiment
```

```python

def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
Sentiment_df['polarity']=Sentiment_df['Brief Review'].apply(polarity)
```

```python
Sentiment_df
```

```python
def review_polarity(pol):
    if pol < 0:
        return "NEGATIVE"
    elif pol > 0:
        return "POSITIVE"
    else:
        return "NEUTRAL"
```

```python
Sentiment_df['Review_Type']=Sentiment_df['polarity'].apply(review_polarity)
```

```python
count_reviews_df=Sentiment_df.groupby('Review_Type',as_index=False)['User ID'].count()
```

```python
count_reviews_df.rename(columns={'User ID':'count of Users'},inplace=True)
```

```python
# Plotting the count of reviews
fig=px.bar(count_reviews_df,x="Review_Type",y='count of Users',color='Review_Type',template="plotly_dark",color_discrete_sequence=['Red','Blue','Green'])
fig.update_layout(width=800,title=dict(text='Sentimental Analysis of Users',x=0.5))
fig.show()
```

```python
# Word Cloud Design
neg_review_df=Sentiment_df[Sentiment_df['Review_Type']=='NEGATIVE']
pos_review_df=Sentiment_df[Sentiment_df['Review_Type']=='POSITIVE']
```

```python
neg_reviews_text=' '.join(neg_review_df['Brief Review'])
pos_reviews_text=' '.join(pos_review_df['Brief Review'])
all_reviews_text=' '.join(Sentiment_df['Brief Review'])
```
